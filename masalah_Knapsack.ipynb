```python
"""
Program Knapsack 0/1
"""
import itertools

def konversi_daftar_pilihanKe_biner(daftarPilihan):
    hasil_konversi_pakai_permutasi_biner_dalam_string = [''.join(seq) for seq in itertools.product('01', repeat=len(daftarPilihan))]
    #print(hasil_konversi_pakai_permutasi_biner)
    hasil_pecahan_tiap_digit_hasil_konversi_permutasi_biner = []
    for i in range(len(hasil_konversi_pakai_permutasi_biner_dalam_string)):
        hasil_pecahan_tiap_digit_hasil_konversi_permutasi_biner.append(list(map(int,str(hasil_konversi_pakai_permutasi_biner_dalam_string[i]))))
    #print(pecah_tiap_digit_hasil_konversi_permutasi_biner)
    return hasil_pecahan_tiap_digit_hasil_konversi_permutasi_biner

def knapsack(daftarPilihan, bobotMaksimal):
    jumlah_bobot = 0
    jumlah_prioritas = 0
    urutan_hasil_PrioritasBobotIndeks = []
    for i in range(len(konversi_daftar_pilihanKe_biner(daftarPilihan))):
        urutan_hasil_PrioritasBobotIndeks.append([])
        for j in range(len(konversi_daftar_pilihanKe_biner(daftarPilihan)[i])):
            print('Indeks-',i,konversi_daftar_pilihanKe_biner(daftarPilihan)[i][j], end='')
            perkalian_tiap_digit_biner_dengan_bobot = konversi_daftar_pilihanKe_biner(daftarPilihan)[i][j] * daftarPilihan[j][0]
            perkalian_tiap_digit_biner_dengan_prioritas = konversi_daftar_pilihanKe_biner(daftarPilihan)[i][j] * daftarPilihan[j][1]
            jumlah_bobot = jumlah_bobot + perkalian_tiap_digit_biner_dengan_bobot
            jumlah_prioritas = jumlah_prioritas + perkalian_tiap_digit_biner_dengan_prioritas
            print(' * bobot',daftarPilihan[j][0],'prioritas',daftarPilihan[j][1],daftarPilihan[j][2],'=',jumlah_bobot)
        selisih_bobotMaksimal_dengan_jumlahBobot = bobotMaksimal - jumlah_bobot
        if selisih_bobotMaksimal_dengan_jumlahBobot <= 0 or jumlah_bobot > bobotMaksimal or jumlah_bobot == 0:
            print('Bobot =',jumlah_bobot,'Ditolak')
        else:
            print('Bobot =',jumlah_bobot,'Diterima | Prioritas',jumlah_prioritas)
            urutan_hasil_PrioritasBobotIndeks[i].append(jumlah_prioritas)
            urutan_hasil_PrioritasBobotIndeks[i].append(jumlah_bobot)
            urutan_hasil_PrioritasBobotIndeks[i].append(i)            
        jumlah_bobot = 0
        jumlah_prioritas = 0
        print()
    print('Item terpilih [Priotas-Bobot-Indeks]=',max(urutan_hasil_PrioritasBobotIndeks))
    urutan_hasil_PrioritasBobotIndeks.sort(reverse=True)
    print(urutan_hasil_PrioritasBobotIndeks)
               
"""
Program utama Knapsack
"""
MAX_WEIGHT = 1000
daftar_pilihan = [
    [150,8, 'ICACSIS'],
    [290,9, 'IcoDSE'],
    [530,11,'SAINS'],
    [380,4, 'ASE']
]

print('Bobot maksimal',MAX_WEIGHT)
konversi_daftar_pilihanKe_biner(daftar_pilihan)
knapsack(daftar_pilihan, MAX_WEIGHT)
```
